### a. One-Sample T-Test


Example 1.Testing if a sample mean is significantly different from a known value

```python
# Biological example: Testing if the average reaction time is different from 0.3 seconds
reaction_times = np.array([0.28, 0.32, 0.29, 0.27, 0.31])

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(reaction_times, 0.3)
print("One-Sample T-Test:")
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Null hypothesis rejected: Average reaction time is significantly different from 0.3 seconds.")
else:
    print("Null hypothesis accepted: Average reaction time is not significantly different from 0.3 seconds.")
```

### b. Pearson Correlation Coefficient


Example 2: Assessing correlation between two sets of measurements



```python
# Neuroscientific example: Correlation between brain volume and cognitive performance scores
brain_volumes = np.array([1200, 1300, 1250, 1100, 1400])  # in cubic centimeters
cognitive_scores = np.array([79, 75, 90, 85, 80])  # cognitive performance scores

# Calculate Pearson correlation coefficient
corr_coef, p_value = stats.pearsonr(brain_volumes, cognitive_scores)
print("\nPearson Correlation Coefficient:")
print("Correlation coefficient:", corr_coef)
print("P-value:", p_value)
if p_value < 0.05:
    print("Correlation is statistically significant: Brain volume correlates with cognitive performance.")
else:
    print("No significant correlation found between brain volume and cognitive performance.")
```

### c. Chi-Square Test


Example 3: Analyzing categorical data



```python
# Sports exercise example: Frequency of exercise preferences among athletes (times per month)
observed = np.array([[30, 20, 32],   # Strength training, Cardio, Yoga
                     [25, 35, 40]])  # start of training  vs close to major competition

# Perform chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)
print("\nChi-Square Test:")
print("Chi2 statistic:", chi2_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Null hypothesis rejected: There is a significant association between exercise preference and training phase.")
else:
    print("Null hypothesis accepted: No significant association found between exercise preference and training phase.")
```

### d. Linear Regression


Example 4: Modeling relationships between variables



```python
# Biological example: Predicting blood pressure based on age
ages = np.array([35, 40, 45, 50, 55])  # Age of individuals (in years)
blood_pressure = np.array([120, 125, 130, 135, 140])  # Blood pressure measurements (mmHg)

# Add constant for intercept in linear regression
ages_const = sm.add_constant(ages)

# Fit linear regression model
model = sm.OLS(blood_pressure, ages_const)
results = model.fit()

# Print regression results
print("\nLinear Regression:")
print(results.summary())
```

### e. Paired T-Test


Example 5: Comparing measurements before and after an intervention



```python
# Neuroscientific example: Comparing cognitive scores before and after brain stimulation
cognitive_scores_before = np.array([75, 80, 78, 70, 82])
cognitive_scores_after = np.array([80, 82, 79, 72, 85])

# Perform paired t-test
t_stat, p_value = stats.ttest_rel(cognitive_scores_after, cognitive_scores_before)
print("\nPaired T-Test:")
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Null hypothesis rejected: There is a significant difference in cognitive scores before and after stimulation.")
else:
    print("Null hypothesis accepted: No significant difference found in cognitive scores before and after stimulation.")
```

```python
# Step 1: Load the Dataset
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Load dataset
neuro_data = pd.read_csv('neuro_data.csv')

# Step 2: Explore the Dataset
# Display first 5 rows
print("First 5 rows of the dataset:")
print(neuro_data.head())

# Check for missing values
print("\nMissing values in the dataset:")
print(neuro_data.isnull().sum())

# Summary statistics
print("\nSummary statistics of firing rates:")
summary_stats = neuro_data[['Pre_Stimulus', 'Post_Stimulus']].describe()
print(summary_stats)

# Step 3: Data Visualization and Statistical Comparison
# Create a bar plot
mean_firing_rates = neuro_data[['Pre_Stimulus', 'Post_Stimulus']].mean()
std_firing_rates = neuro_data[['Pre_Stimulus', 'Post_Stimulus']].std()
labels = ['Pre Stimulus', 'Post Stimulus']

plt.bar(labels, mean_firing_rates, yerr=std_firing_rates, capsize=5)
plt.ylabel('Mean Firing Rates (spikes per second)')
plt.title('Comparison of Firing Rates Before and After Stimulus')
plt.show()

# Perform paired t-test
t_stat, p_value = ttest_rel(neuro_data['Pre_Stimulus'], neuro_data['Post_Stimulus'])

# Interpret the result
print("\nPaired t-test results:")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("There is a significant difference in firing rates before and after the stimulus.")
else:
    print("There is no significant difference in firing rates before and after the stimulus.")

# Step 4 (Optional): Advanced Analysis
# Compute correlation matrix
correlation_matrix = neuro_data[['Pre_Stimulus', 'Post_Stimulus']].corr()

# Visualize correlation matrix as heatmap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Matrix of Firing Rates')
plt.show()
```
